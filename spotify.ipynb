Top 50 songs listened in 2021 in Spotify

```python
import csv
import seaborn as srn
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
#importar dados
dataset = pd.read_csv('spotify_top50_2021.csv', sep=",")
#visualizar
dataset.head()
```

```python
#tamanho
dataset.shape
```

```python
#visualizando quantidades
agrupado = dataset.groupby(['popularity', 'artist_name', 'track_name']).size()
agrupado
```

```python
#popularidade
srn.displot(dataset.popularity, bins = 10)
```

```python
#tempo
srn.displot(dataset.tempo, bins = 10)
```

```python
#loudness
srn.displot(dataset['loudness'])
```

```python
dataset.plot(x="artist_name", y=["popularity"], kind="bar", figsize=(20,5))
```

```python
dataset.plot(x="id", y=["danceability", "energy"], kind="bar", figsize=(20,5)) 
```

```python
dataset.plot(x="id", y=["speechiness","acousticness","instrumentalness"], kind="bar", figsize=(20,5)) 
```

```python
dataset.plot(x="id", y=["liveness","valence"], kind="bar", figsize=(20,5)) 
```
